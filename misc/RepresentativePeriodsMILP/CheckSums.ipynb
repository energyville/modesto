```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('results_ordered/1dnewsol/result_ordered12p.txt', sep=' ', index_col=0)
```

```python
data
```

```python
data['Full_dem'] = data[['E_demand_full', 'E_loss_stor_full', 'E_net_loss_full']].sum(axis=1)
```

```python
data['Full_prod'] = data[['E_sol_full', 'E_backup_full']].sum(axis=1)
```

```python
data['Repr_dem'] = data[['E_demand_repr', 'E_loss_stor_repr', 'E_net_loss_repr']].sum(axis=1)
data['Repr_prod'] = data[['E_sol_repr', 'E_backup_repr']].sum(axis=1)
```

```python
%matplotlib notebook

offset = 2500

plt.plot(data['A'], data['Full_dem'], 'o', label='Full dem side')
plt.plot(data['A']+offset, data['Full_prod'], 'o', label='Full prod side')
plt.plot(data['A'], data['Repr_dem'], 's', label='Repr dem side')
plt.plot(data['A']+offset, data['Repr_prod'], 's', label='Repr prod side')
plt.hlines([data['Full_dem'].mean()*0.975, data['Full_dem'].mean()*1.025], xmin=20000, xmax=80000, linestyles=':')
plt.plot(data['A'], data['E_demand_full'], '^', label='Full demand')
plt.plot(data['A'], data['E_demand_repr'], '^', label='Repr demand')
plt.plot(data['A']+offset, data['E_curt_repr'], )
plt.legend()
plt.ylim(0,None)
```

```python
%matplotlib notebook 

name = 'backup'
fig, ax = plt.subplots()
for i, row in data.iterrows():
    ax.add_line(plt.Line2D(xdata=[row['A'], row['A']+2500], ydata=[row['E_{}_full'.format(name)], 
              row['E_{}_repr'.format(name)]], linestyle='-', linewidth=0.5))
    
ax.plot(data['A'], data['E_{}_full'.format(name)], 'o', label='Full {}'.format(name))
ax.plot(data['A']+2500, data['E_{}_repr'.format(name)], 's', label='Repr {}'.format(name))
ax.legend()
ax.set_ylim(0, None)
```

```python
%matplotlib notebook 

name = 'demand'
fig, ax = plt.subplots()
for i, row in data.iterrows():
    ax.add_line(plt.Line2D(xdata=[row['A'], row['A']+2500], ydata=[row['E_{}_full'.format(name)], 
              row['E_{}_repr'.format(name)]], linestyle='-', linewidth=0.5))
ax.hlines(0.975*data['E_{}_full'.format(name)].iloc[0], xmin=0, xmax=110000, linestyles=':')
ax.plot(data['A'], data['E_{}_full'.format(name)], 'o', label='Full {}'.format(name))
ax.plot(data['A']+2500, data['E_{}_repr'.format(name)], 's', label='Repr {}'.format(name))
ax.legend()
ax.set_ylim(0, None)
```
