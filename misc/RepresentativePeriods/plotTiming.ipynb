```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_pickle('timing_df.pkl')
```

```python
import numpy as np
print np.array([6, 8, 10, 12])*3
print np.array([6, 8, 10, 12])*7

def map_fun(x):
    out = []
    for i in x:
        if i in [18, 24, 30, 36]:
            out.append(3) 
        elif i in [42, 56, 70, 84]:
            out.append(7)
        else:
            out.append(1)
    return out
```

```python
df['daylen'] = map_fun(df['days'])
```
